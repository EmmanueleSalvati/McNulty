```python
import McNulty_helper
from McNulty_read_sql import import_df

raw_df = import_df('all_hospitals')
raw_df = McNulty_helper.clean_question_marks(raw_df)
McNulty_helper.zero_to_NaN(raw_df, 'chol_mg_dl')
McNulty_helper.zero_to_NaN(raw_df, 'rest_bp')
McNulty_helper.change_types(raw_df)
df = McNulty_helper.NaN_to_modes(raw_df)
df = McNulty_helper.reduce_diagnosis(df)

%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

men = df.loc[df['sex']==1.0,]
```

```python
men.columns.values.tolist()
```

```python
g = sns.FacetGrid(men, col="hospital", sharey=False, size=4, aspect=1.)
g.map(sns.barplot, 'rest_ecg')
g.set_xlabels('ECG at rest')
```

```python
g = sns.FacetGrid(men, col="hospital", sharey=False, size=4, aspect=1.)
g.map(sns.barplot, 'st_exercise_angina')
g.set_xlabels('exercise-induced angina')
```

```python
g = sns.FacetGrid(men, col="hospital", sharey=False, size=4, aspect=1.)
g.map(sns.barplot, 'st_exercise_slope')
g.set_xlabels('exercised-induced slope')
```

```python
g = sns.FacetGrid(men, col="hospital", sharey=False, size=4, aspect=1.)
g.map(plt.hist, 'st_max_heart_rt_ach', histtype='bar', align='mid')
g.set_xlabels('ST max heart rate')
```

```python
g = sns.FacetGrid(men, col='hospital', sharey=False, size=4, aspect=1.)
g.map(plt.hist, 'st_depression', histtype='bar', align='mid')
g.set_xlabels('ST depression')
```

```python
g = sns.FacetGrid(men, col='hospital', sharey=False, sharex=False, size=4, aspect=1.)
g.map(sns.barplot, 'colored_vessels')
g.set_xlabels('# major vessels colored fluoroscopy')
```
